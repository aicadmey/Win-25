```python
# example of a simple Multi-Layer Perceptron (MLP) implemented using Python with the TensorFlow/Keras library.
# This MLP includes three hidden layers,each with a different number of nodes.


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the MLP model
model = Sequential([
    Dense(64, activation='relu',input_shape=(input_dim,)),  # First hidden layer with 64 nodes
    Dense(32, activation='relu'),                            # Second hidden layer with 32 nodes
    Dense(16, activation='relu'),                            # Third hidden layer with 16 nodes
    Dense(output_dim, activation='softmax')                  # Output layer
    ])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary
model.summary()

```
