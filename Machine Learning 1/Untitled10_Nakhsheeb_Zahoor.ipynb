```python
import numpy as np
import matplotlib.pyplot as plt

#  Input data
# Example data (Hours studied vs Marks obtained)
X = np.array([1, 2, 3, 4, 5])  # Independent variable (Hours studied)
Y = np.array([20, 30, 50, 87, 100])  # Dependent variable (Marks obtained)

#  Calculate the required sums
n = len(X)  # Number of data points
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X ** 2)

#  Calculate slope (m) and intercept (b)
m = (n * sum_XY - sum_X * sum_Y) / (n * sum_X2 - sum_X ** 2)
b = (sum_Y - m * sum_X) / n

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

#  Write the regression line equation
def predict(x):
    return m * x + b

#  Predict values
Y_predicted = predict(X)

print("Predicted Y values:", Y_predicted)

# Visualize the data and the regression line
plt.scatter(X, Y, color='green', label='Actual Data')  # Scatter plot for actual data
plt.plot(X, Y_predicted, color='black', label='Regression Line')  # Line for predicted values
plt.xlabel("Hours Studied (X)")
plt.ylabel("Marks Obtained (Y)")
plt.title("Linear Regression Example")
plt.legend()
plt.show()
```
