```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate synthetic data (binary classification)
X, y = make_classification(n_samples=100, n_features=50, n_classes=2, random_state=42)

# Train logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plotting the data and the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.title('Logistic Regression on Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

print(f"Model Coefficients: {model.coef_}")
```
