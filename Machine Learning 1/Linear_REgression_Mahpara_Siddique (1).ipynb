**Space:** Signal Delay vs Distance of Planets
**Problem:** Investigate how signal delay increases with distance from Earth.

Experiment1

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([50, 100, 150, 200, 250])  # Distance from Earth in millions of km
Y = np.array([2.5, 5.0, 7.5, 10.0, 12.5])  # Signal delay in minutes

# Linear Regression Calculation
n = len(X)
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X ** 2)

# Calculate slope (m) and intercept (b)
m = (n * sum_XY - sum_X * sum_Y) / (n * sum_X2 - sum_X ** 2)
b = (sum_Y - m * sum_X) / n

# Predictions
Y_pred = m * X + b

# Output results
print("Slope (m):", m)
print("Intercept (b):", b)
print("Predicted Y values:", Y_pred)

# Plot
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title("Signal Delay vs Distance")
plt.xlabel("Distance from Earth (million km)")
plt.ylabel("Signal Delay (minutes)")
plt.legend()
plt.show()
```

**Alien Life:** Activity Levels vs Temperature
**Problem:** Study how alien activity levels vary with environmental temperature.

Experiment5

```python
import numpy as np
import matplotlib.pyplot as plt


# Dataset
X = np.array([-100, -50, 0, 50, 100])  # Temperature in degrees Celsius
Y = np.array([0, 10, 25, 10, 0])  # Alien activity levels (arbitrary units)

# Since this dataset is non-linear, only use the increasing portion for regression
X_linear = X[:3]  # Take -100, -50, and 0
Y_linear = Y[:3]

# Linear Regression Calculation
n = len(X_linear)
sum_X = np.sum(X_linear)
sum_Y = np.sum(Y_linear)
sum_XY = np.sum(X_linear * Y_linear)
sum_X2 = np.sum(X_linear ** 2)

# Calculate slope (m) and intercept (b)
m = (n * sum_XY - sum_X * sum_Y) / (n * sum_X2 - sum_X ** 2)
b = (sum_Y - m * sum_X) / n

# Predictions
Y_pred = m * X_linear + b

# Output results
print("Slope (m):", m)
print("Intercept (b):", b)
print("Predicted Y values:", Y_pred)

# Plot
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X_linear, Y_pred, color='red', label='Regression Line (Subset Data)')
plt.title("Alien Activity vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Alien Activity Levels")
plt.legend()
plt.show()
```
