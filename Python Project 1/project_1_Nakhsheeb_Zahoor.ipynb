```python
# Initialize traits with default values
Traits = {
    "Self sufficient": 0,
    "Seek approval": 0
}

# Define the questions and options
questions = [
    {
        "question": "Would you rather be:",
        "options": [
            "A. Smartest person around.",
            "B. The most charming."
        ],
        "choices": [
            ("Self sufficient", "A"),
            ("Seek approval", "B")
        ]
    }
]

# Function to ask questions
def ask_question(question_data):
    print("\n" + question_data["question"])
    for option in question_data["options"]:
        print(option)
    while True:
        choice = input("\nChoose an option (A, B): ").upper()
        if choice in ['A', 'B']:
            return choice
        else:
            print("Invalid choice. Please choose either A or B.")

# Function to check answers and update traits
def check_answer(user_choice, choices):
    for trait, option in choices:
        if user_choice == option:
            Traits[trait] += 1

# Display welcome message
print("Welcome to the Psychological Quiz!")
print(
    "I am not a psychologist nor a psychiatrist, "
    "just a random person interested in psychology. "
    "These are my assumptions and may not be correct, "
    "so don't take it to heart. Thanks!"
)

# Ask questions and update traits
for question_data in questions:
    user_choice = ask_question(question_data)
    check_answer(user_choice, question_data["choices"])

# Calculate and display results
print("\nCalculating your results...\n")
if Traits["Self sufficient"] > Traits["Seek approval"]:
    print("You are: Self sufficient!")
    print("It means you don't care about other people's opinions, and you are happy with yourself.")
else:
    print("You need approval.")
    print("You seek approval from others, which is common nowadays but can be straining.")



```
