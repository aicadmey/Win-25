```python

# Example 6 print number only primary number
num = 69
is_prime = True
for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
if is_prime:
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

● 10 Example of while loop::

```python

# Example 1
while True:
    print("sorry😐")
    if input("Do you want to stop? (yes/no): ") == "yes":
        break
```

```python

# Example 2
count = 5
while count <= 100:
    print(count)
    count += 20
```

```python

# Example 3
number = 1
sum = 0
while number <= 69:
    sum += number
    number += 1
print("Sum of numbers from 1 to 69:", sum)
```

```python

# Example 4
number = int(input("Enter a number: "))
divisor = 2
is_prime = True
while divisor < number:
    if number % divisor == 0:
        is_prime = False
        break
    divisor += 1
if is_prime:
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

```python

# Example 5
string = "Davil World 😈"
for char in string:
    print(char)
```

```python

#Example 6
while True:
    age = input("Enter your age (positive integer): ")
    if age.isdigit() and int(age) > 0:
        break
    else:
        print("Invalid input. Please enter a positive integer.")
print("Your age is:", age)
```

```python

# Example 7
import random
while True:
    coin = random.randint(0, 1)
    if coin == 0:
        print("Heads")
    else:
        print("Tails")
    if input("Do you want to toss again? (yes/no): ") == "no":
        break
```

```python

# Example 3 + 5 = 8
count = 1
while count <= 50:
    print(count)
    count += 1
```

```python

# Example 9
number = int(input("Enter a number: "))
factorial = 1
while number > 0:
    factorial *= number
    number -= 1
print("Factorial:", factorial)
```

```python

# Example 9
n = int(input("Enter a number: "))
sum = 0
i = 1
while i <= n:
    sum += i
    i += 1
print("Sum of numbers:", sum)
```

```python

# Example 10
while True:
    num = int(input("Enter a positive number: "))
    if num > 0:
        break
    print("Invalid input. Please enter a positive number.")
```

```python

# Example 11
my_list = [1, 2, 3, 4, 5]
i = 0
while i < len(my_list):
    print(my_list[i])
    i += 1
```
