```python
# Base class representing a general celestial body
class CelestialBody:
    def __init__(self, name, mass, diameter):
        self.name = name
        self.mass = mass
        self.diameter = diameter

    def describe(self):
        return f"{self.name} is a celestial body with a mass of {self.mass} kg and a diameter of {self.diameter} km."

# Subclass representing a planet
class Planet(CelestialBody):
    def __init__(self, name, mass, diameter, moons_count):
        super().__init__(name, mass, diameter)
        self.moons_count = moons_count

    def describe(self):
        description = super().describe()
        return f"{description} It has {self.moons_count} moons."


earth = Planet("Earth", 5.972e24, 12742, 1)


print(earth.describe())

# 2
# Base class
class CelestialBody:
    def __init__(self, name, mass, diameter):
        self.name = name
        self.mass = mass
        self.diameter = diameter

    def describe(self):
        return f"{self.name} is a celestial body with a mass of {self.mass} kg and a diameter of {self.diameter} km."


class Star(CelestialBody):
    def __init__(self, name, mass, diameter, temperature):
        super().__init__(name, mass, diameter)
        self.temperature = temperature

    def describe(self):
        description = super().describe()
        return f"{description} It has a surface temperature of {self.temperature}K."

class Planet(CelestialBody):
    def __init__(self, name, mass, diameter, moons_count):
        super().__init__(name, mass, diameter)
        self.moons_count = moons_count

    def describe(self):
        description = super().describe()
        return f"{description} It has {self.moons_count} moons."


class Supernova(Star):
    def __init__(self, name, mass, diameter, temperature, explosion_energy):
        super().__init__(name, mass, diameter, temperature)
        self.explosion_energy = explosion_energy

    def describe(self):
        description = super().describe()
        return f"{description} It is a supernova with an explosion energy of {self.explosion_energy} joules."


earth = Planet("Earth", 5.972e24, 12742, 1)
sun = Star("Sun", 1.989e30, 1392700, 5778)
supernova_example = Supernova("SN 1987A", 1.5e32, 100000, 10000000, 1.0e44)


print(earth.describe())
print(sun.describe())
print(supernova_example.describe())
# 333

class LifeForm:
    def __init__(self, name, habitat, lifespan):
        self.name = name
        self.habitat = habitat
        self.lifespan = lifespan

    def evolve(self):
        return f"{self.name} is evolving."

    def describe(self):
        return f"{self.name} lives in {self.habitat} and has a lifespan of {self.lifespan} years."

class Plant(LifeForm):
    def __init__(self, name, habitat, lifespan, photosynthesis_rate):
        super().__init__(name, habitat, lifespan)
        self.photosynthesis_rate = photosynthesis_rate

    def evolve(self):
        return f"{self.name} is evolving to better adapt to its environment for efficient photosynthesis."

    def describe(self):
        description = super().describe()
        return f"{description} It performs photosynthesis at a rate of {self.photosynthesis_rate} grams per day."


class Animal(LifeForm):
    def __init__(self, name, habitat, lifespan, diet):
        super().__init__(name, habitat, lifespan)
        self.diet = diet

    def evolve(self):
        return f"{self.name} is evolving to improve its survival chances through adaptation to the environment."

    def describe(self):
        description = super().describe()
        return f"{description} It has a {self.diet} diet."


class Human(Animal):
    def __init__(self, name, habitat, lifespan, diet, intelligence_level):
        super().__init__(name, habitat, lifespan, diet)
        self.intelligence_level = intelligence_level

    def evolve(self):
        return f"{self.name} is evolving intellectually to better manipulate its environment."

    def describe(self):
        description = super().describe()
        return f"{description} Intelligence level: {self.intelligence_level}."


class Bird(Animal):
    def __init__(self, name, habitat, lifespan, diet, wing_span):
        super().__init__(name, habitat, lifespan, diet)
        self.wing_span = wing_span

    def evolve(self):
        return f"{self.name} is evolving for improved flight efficiency and migration patterns."

    def describe(self):
        description = super().describe()
        return f"{description} It has a wingspan of {self.wing_span} meters."


oak_tree = Plant("Oak Tree", "forest", 300, 2.5)
lion = Animal("Lion", "savannah", 14, "carnivore")
human = Human("Homo sapiens", "varied", 80, "omnivore", 130)
human = Human("Homo sapiens", "varied", 80, "omnivore", 130)
eagle = Bird("Eagle", "mountains", 20, "carnivore", 2.50)
print(oak_tree.describe())
print(oak_tree.evolve())

print(lion.describe())
print(lion.evolve())

print(human.describe())
print(human.evolve())
#333

class LifeForm:
    def __init__(self, name, habitat, lifespan):
        self.name = name
        self.habitat = habitat
        self.lifespan = lifespan

    def evolve(self):
        return f"{self.name} is evolving."

    def describe(self):
        return f"{self.name} lives in {self.habitat} and has a lifespan of {self.lifespan} years."


class Plant(LifeForm):
    def __init__(self, name, habitat, lifespan, photosynthesis_rate):
        super().__init__(name, habitat, lifespan)
        self.photosynthesis_rate = photosynthesis_rate

    def evolve(self):
        return f"{self.name} is evolving to increase its photosynthesis efficiency."

    def describe(self):
        description = super().describe()
        return f"{description} It performs photosynthesis at a rate of {self.photosynthesis_rate} grams per day."


class Animal(LifeForm):
    def __init__(self, name, habitat, lifespan, diet):
        super().__init__(name, habitat, lifespan)
        self.diet = diet

    def evolve(self):
        return f"{self.name} is evolving to improve its survival skills and adaptation to the environment."

    def describe(self):
        description = super().describe()
        return f"{description} It has a {self.diet} diet."


class Human(Animal):
    def __init__(self, name, habitat, lifespan, diet, intelligence_level):
        super().__init__(name, habitat, lifespan, diet)
        self.intelligence_level = intelligence_level

    def evolve(self):
        return f"{self.name} is evolving intellectually, mastering complex tools and social systems."

    def describe(self):
        description = super().describe()
        return f"{description} Intelligence level: {self.intelligence_level}."


class Bird(Animal):
    def __init__(self, name, habitat, lifespan, diet, wing_span):
        super().__init__(name, habitat, lifespan, diet)
        self.wing_span = wing_span

    def evolve(self):
        return f"{self.name} is evolving for improved flight and migration abilities."

    def describe(self):
        description = super().describe()
        return f"{description} It has a wingspan of {self.wing_span} meters."


class AquaticLife(Animal):
    def __init__(self, name, habitat, lifespan, diet, swim_speed):
        super().__init__(name, habitat, lifespan, diet)
        self.swim_speed = swim_speed

    def evolve(self):
        return f"{self.name} is evolving to increase its speed and survival in the water."

    def describe(self):
        description = super().describe()
        return f"{description} It swims at a speed of {self.swim_speed} km/h."


class Reptile(Animal):
    def __init__(self, name, habitat, lifespan, diet, cold_blooded=True):
        super().__init__(name, habitat, lifespan, diet)
        self.cold_blooded = cold_blooded

    def evolve(self):
        return f"{self.name} is evolving to adapt to extreme temperatures and diverse environments."

    def describe(self):
        description = super().describe()
        return f"{description} It is {'cold-blooded' if self.cold_blooded else 'warm-blooded'}."


oak_tree = Plant("Oak Tree", "forest", 300, 3.0)
lion = Animal("Lion", "savannah", 14, "carnivore")
human = Human("Homo sapiens", "varied", 80, "omnivore", 130)
eagle = Bird("Eagle", "mountains", 20, "carnivore", 2.5)
shark = AquaticLife("Shark", "ocean", 30, "carnivore", 60)
crocodile = Reptile("Crocodile", "swamps", 70, "carnivore")


print(oak_tree.describe())
print(oak_tree.evolve())

print(lion.describe())
print(lion.evolve())

print(human.describe())
print(human.evolve())

print(eagle.describe())
print(eagle.evolve())

print(shark.describe())
print(shark.evolve())

print(crocodile.describe())
print(crocodile.evolve())



```
