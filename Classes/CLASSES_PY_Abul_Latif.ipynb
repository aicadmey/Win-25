```python
class Species:
    def __init__(self, name, environment):
        self.name = name
        self.environment = environment

    def adapt(self):
        if self.environment == "desert":
            return f"{self.name} grows thicker skin to conserve water."
        elif self.environment == "forest":
            return f"{self.name} develops better camouflage in the forest."
        else:
            return f"{self.name} adapts to the {self.environment} environment."


lion = Species("Lion", "forest")
camel = Species("Camel", "desert")


print(lion.adapt())
print(camel.adapt())
# SECOND EXAMPLES

class Mutation:
    def __init__(self, species, mutation_type):
        self.species = species
        self.mutation_type = mutation_type

    def show_mutation(self):
        return f"The {self.species} has undergone a {self.mutation_type} mutation."


eagle = Mutation("Eagle", "feather color change")
fish = Mutation("Fish", "increased resistance to cold")

print(eagle.show_mutation())
print(fish.show_mutation())
#THIRD EXAMPLES

class NaturalSelection:
    def __init__(self, species, fitness_level):
        self.species = species
        self.fitness_level = fitness_level

    def survival_chance(self):
        if self.fitness_level > 8:
            return f"{self.species} has a high chance of survival due to strong fitness."
        else:
            return f"{self.species} may struggle to survive due to low fitness."


cheetah = NaturalSelection("Cheetah", 9)
sloth = NaturalSelection("Sloth", 4)
print(cheetah.survival_chance())
print(sloth.survival_chance())

#FOURT EXAMPLE
class Speciation:
    def __init__(self, original_species, new_species):
        self.original_species = original_species
        self.new_species = new_species

    def describe_speciation(self):
        return f"Due to isolation, {self.original_species} has evolved into a new species: {self.new_species}."


wolf = Speciation("Wolf", "Dingo")
finch = Speciation("Finch", "Galapagos Finch")
print(wolf.describe_speciation())
print(finch.describe_speciation())

#FIFTH EXAMPLES
class Planet:
    def __init__(self, name, size, type_):
        self.name = name
        self.size = size
        self.type_ = type_

    def planet_info(self):
        return f"{self.name} is a {self.type_} planet with a size of {self.size}."


earth = Planet("Earth", "12,742 km", "terrestrial")
jupiter = Planet("Jupiter", "139,820 km", "gas giant")

print(earth.planet_info())
print(jupiter.planet_info())
#SIXTH EXAMPLE
class Star:
    def __init__(self, name, temperature, size):
        self.name = name
        self.temperature = temperature
        self.size = size

    def star_type(self):
        if self.temperature > 10000:
            return f"{self.name} is a hot star with a temperature of {self.temperature} K."
        else:
            return f"{self.name} is a cool star with a temperature of {self.temperature} K."

sun = Star("Sun", 5778, "medium")
sirius = Star("Sirius", 9940, "large")


print(sun.star_type())
print(sirius.star_type())
#SEVENTH EXAMPLE
class Spacecraft:
    def __init__(self, name, mission, destination):
        self.name = name
        self.mission = mission
        self.destination = destination

    def mission_info(self):
        return f"{self.name} is on a {self.mission} mission to {self.destination}."

voyager = Spacecraft("Voyager 1", "exploration", "interstellar space")
curiosity = Spacecraft("Curiosity Rover", "research", "Mars")


print(voyager.mission_info())
print(curiosity.mission_info())
#EIGHTTH
class Galaxy:
    def __init__(self, name, type_, size):
        self.name = name
        self.type_ = type_
        self.size = size

    def galaxy_info(self):
        return f"{self.name} is a {self.type_} galaxy with a size of {self.size} light-years."

milky_way = Galaxy("Milky Way", "spiral", "100,000")
andromeda = Galaxy("Andromeda", "spiral", "220,000")


print(milky_way.galaxy_info())
print(andromeda.galaxy_info())
#NINTH EXAMPLES
class RobotInteraction:
    def __init__(self, robot_name):
        self.robot_name = robot_name

    def respond_to_command(self, command):
        return f"{self.robot_name} is responding: {command}"


assistant_robot = RobotInteraction("AssistBot")
service_robot = RobotInteraction("ServeBot")


print(assistant_robot.respond_to_command("How can I assist you?"))
print(service_robot.respond_to_command("Please bring me a drink."))
#TENTH EXAMPLE
class RobotFunctionality:
    def __init__(self, robot_name, task):
        self.robot_name = robot_name
        self.task = task

    def perform_task(self):
        return f"{self.robot_name} is performing the task: {self.task}."


welding_robot = RobotFunctionality("WeldBot", "welding metal")
painting_robot = RobotFunctionality("PaintBot", "painting surfaces")
print(welding_robot.perform_task())
print(painting_robot.perform_task())
#ELEVENTH EXAMPLES
class RobotMovement:
    def __init__(self, robot_name, speed):
        self.robot_name = robot_name
        self.speed = speed

    def move(self, direction):
        return f"{self.robot_name} moves {direction} at {self.speed} speed."


rover = RobotMovement("Rover", "fast")
cleaner = RobotMovement("CleanerBot", "slow")


print(rover.move("forward"))
print(cleaner.move("backward"))










```
