```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("animal sound")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        print(f"Woof! I'm a {self.breed} named {self.name}")

# Create a Dog object and call its methods
my_dog = Dog("kutta", "dog")
my_dog.speak()
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        pass

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def speak(self):
        print("Woof!")

class Cat(Animal):
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color

    def speak(self):
        print("Meow!")


# Create instances of the subclasses
dog1 = Dog("kutta", 3, "German")
cat1 = Cat("billi", 2, "billaa")


# Print the values of the instances
print(dog1.name, dog1.age, dog1.breed)
print(cat1.name, cat1.age, cat1.color)
```

```python
class Shape:
    def __init__(self, color):
        self.color = color

    def print_info(self):
        print(f"Color: {self.color}")


class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def print_info(self):
        super().print_info()
        print(f"Radius: {self.radius} units")


class Square(Shape):
    def __init__(self, color, side_length):
        super().__init__(color)
        self.side_length = side_length

    def print_info(self):
        super().print_info()
        print(f"Side Length: {self.side_length} units")


class Triangle(Shape):
    def __init__(self, color, base, height):
        super().__init__(color)
        self.base = base
        self.height = height

    def print_info(self):
        super().print_info()
        print(f"Base: {self.base} units")
        print(f"Height: {self.height} units")


# Creating objects and printing information
circle1 = Circle("Red", 5)
circle1.print_info()

square1 = Square("Blue", 10)
square1.print_info()

triangle1 = Triangle("Green", 6, 8)
triangle1.print_info()
```

```python
class Vehicle:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def print_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")


class Car(Vehicle):
    def __init__(self, brand, model, year, num_doors, fuel_type):
        super().__init__(brand, model, year)
        self.num_doors = num_doors
        self.fuel_type = fuel_type

    def print_info(self):
        super().print_info()
        print(f"Number of Doors: {self.num_doors}")
        print(f"Fuel Type: {self.fuel_type}")


class Truck(Vehicle):
    def __init__(self, brand, model, year, payload_capacity, drive_type):
        super().__init__(brand, model, year)
        self.payload_capacity = payload_capacity
        self.drive_type = drive_type

    def print_info(self):
        super().print_info()
        print(f"Payload Capacity: {self.payload_capacity} lbs")
        print(f"Drive Type: {self.drive_type}")


class Motorcycle(Vehicle):
    def __init__(self, brand, model, year, engine_size, seat_height):
        super().__init__(brand, model, year)
        self.engine_size = engine_size
        self.seat_height = seat_height

    def print_info(self):
        super().print_info()
        print(f"Engine Size: {self.engine_size} cc")
        print(f"Seat Height: {self.seat_height} inches")


class ElectricCar(Car):
    def __init__(self, brand, model, year, num_doors, battery_range):
        super().__init__(brand, model, year, num_doors, "Electric")
        self.battery_range = battery_range

    def print_info(self):
        super().print_info()
        print(f"Battery Range: {self.battery_range} miles")


class DieselTruck(Truck):
    def __init__(self, brand, model, year, payload_capacity, fuel_tank_capacity):
        super().__init__(brand, model, year, payload_capacity, "Diesel")
        self.fuel_tank_capacity = fuel_tank_capacity

    def print_info(self):
        super().print_info()
        print(f"Fuel Tank Capacity: {self.fuel_tank_capacity} gallons")


class SportMotorcycle(Motorcycle):
    def __init__(self, brand, model, year, engine_size, horsepower):
        super().__init__(self, brand, model, year, engine_size)
        self.horsepower = horsepower

    def print_info(self):
        super().print_info()
        print(f"Horsepower: {self.horsepower} hp")


class HybridCar(Car):
    def __init__(self, brand, model, year, num_doors, fuel_economy):
        super().__init__(brand, model, year, num_doors, "Hybrid")
        self.fuel_economy = fuel_economy

    def print_info(self):
        super().print_info()
        print(f"Fuel Economy: {self.fuel_economy} mpg")

car1 = ElectricCar("Tesla", "Model 3", 2023, 4, 350)
car1.print_info()

truck1 = DieselTruck("Ford", "F-250", 2022, 3000, 36)
truck1.print_info()

motorcycle1 = SportMotorcycle("Yamaha", "R1", 2024, 1000, 200)
motorcycle1.print_info()

car2 = HybridCar("Toyota", "Prius", 2023, 4, 58)
car2.print_info()
```

```python
class Food:
    def __init__(self, name, calories):
        self.name = name
        self.calories = calories

    def print_info(self):
        print(f"Name: {self.name}")
        print(f"Calories: {self.calories}")


class Fruit(Food):
    def __init__(self, name, calories, color):
        super().__init__(name, calories)
        self.color = color

    def print_info(self):
        super().print_info()
        print(f"Color: {self.color}")


class Vegetable(Food):
    def __init__(self, name, calories, vitamin_rich):
        super().__init__(name, calories)
        self.vitamin_rich = vitamin_rich

    def print_info(self):
        super().print_info()
        print(f"Vitamin Rich: {self.vitamin_rich}")


class DairyProduct(Food):
    def __init__(self, name, calories, fat_content):
        super().__init__(name, calories)
        self.fat_content = fat_content

    def print_info(self):
        super().print_info()
        print(f"Fat Content: {self.fat_content}%")


class Meat(Food):
    def __init__(self, name, calories, protein_content):
        super().__init__(name, calories)
        self.protein_content = protein_content

    def print_info(self):
        super().print_info()
        print(f"Protein Content: {self.protein_content}%")


# Subclasses of Fruit
class CitrusFruit(Fruit):
    def __init__(self, name, calories, color, citrus_acid):
        super().__init__(name, calories, color)
        self.citrus_acid = citrus_acid

    def print_info(self):
        super().print_info()
        print(f"Citrus Acid: {self.citrus_acid}")


class Berry(Fruit):
    def __init__(self, name, calories, color, antioxidant_rich):
        super().__init__(name, calories, color)
        self.antioxidant_rich = antioxidant_rich

    def print_info(self):
        super().print_info()
        print(f"Antioxidant Rich: {self.antioxidant_rich}")


class TropicalFruit(Fruit):
    def __init__(self, name, calories, color, exotic_flavor):
        super().__init__(name, calories, color)
        self.exotic_flavor = exotic_flavor

    def print_info(self):
        super().print_info()
        print(f"Exotic Flavor: {self.exotic_flavor}")


class StoneFruit(Fruit):
    def __init__(self, name, calories, color, pit_type):
        super().__init__(name, calories, color)
        self.pit_type = pit_type

    def print_info(self):
        super().print_info()
        print(f"Pit Type: {self.pit_type}")


# Subclasses of Vegetable
class LeafyGreen(Vegetable):
    def __init__(self, name, calories, vitamin_rich, fiber_content):
        super().__init__(name, calories, vitamin_rich)
        self.fiber_content = fiber_content

    def print_info(self):
        super().print_info()
        print(f"Fiber Content: {self.fiber_content} grams")


class RootVegetable(Vegetable):
    def __init__(self, name, calories, vitamin_rich, underground_part):
        super().__init__(name, calories, vitamin_rich)
        self.underground_part = underground_part

    def print_info(self):
        super().print_info()
        print(f"Underground Part: {self.underground_part}")


class CruciferousVegetable(Vegetable):
    def __init__(self, name, calories, vitamin_rich, sulfur_content):
        super().__init__(name, calories, vitamin_rich)
        self.sulfur_content = sulfur_content

    def print_info(self):
        super().print_info()
        print(f"Sulfur Content: {self.sulfur_content}")


class AlliumVegetable(Vegetable):
    def __init__(self, name, calories, vitamin_rich, pungent_flavor):
        super().__init__(name, calories, vitamin_rich)
        self.pungent_flavor = pungent_flavor

    def print_info(self):
        super().print_info()
        print(f"Pungent Flavor: {self.pungent_flavor}")
```
