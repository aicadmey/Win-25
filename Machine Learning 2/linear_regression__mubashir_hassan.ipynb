```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))  # Reshape for sklearn
y = np.array([2, 4, 5, 4, 5])
model = LinearRegression().fit(X, y)
print("Example 1: Intercept:", model.intercept_, "Coefficient:", model.coef_)
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.title("Example 1")
plt.show()
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))
y = np.array([3, 6, 9, 12, 15])
model = LinearRegression().fit(X, y)
print("Example 2: Intercept:", model.intercept_, "Coefficient:", model.coef_)
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.title("Example 2")
plt.show()
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))
y = np.array([2, 4, 5, 4, 5])
model = LinearRegression().fit(X, y)
new_x = np.array([[6]])  # Predict for x = 6
prediction = model.predict(new_x)
print(f"Example 3: Prediction for x = 6: {prediction[0]:.2f}")
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))
y = np.array([3, 6, 9, 12, 15])
model = LinearRegression().fit(X, y)
new_X = np.array([[6], [7], [8]])
predictions = model.predict(new_X)
print(f"Example 4: Predictions for x = 6, 7, 8: {predictions}")
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))
y = np.array([10, 8, 6, 4, 2])
model = LinearRegression().fit(X, y)
intercept = model.intercept_
coefficient = model.coef_[0]
print(f"Example 5: Equation: y = {coefficient:.2f}x + {intercept:.2f}")
```
