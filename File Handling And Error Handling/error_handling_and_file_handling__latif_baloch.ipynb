```python
def validate_coordinates(coordinate):
    try:
        # Try to convert the coordinate to a float (assuming it's a valid number)
        coord_value = float(coordinate)
        if coord_value < -90 or coord_value > 90:
            raise ValueError("Coordinate must be between -90 and 90 degrees.")
        return coord_value
    except ValueError as e:
        print(f"Invalid input for coordinate: {e}")

# Example of usage
validate_coordinates("100")  # This will trigger the error

#2nd
space_missions = {
    "Mars Rover": {"launch_year": 2020, "status": "active"},
    "Hubble Telescope": {"launch_year": 1990, "status": "active"},
}

def get_mission_data(mission_name):
    try:
        mission_data = space_missions[mission_name]
        return mission_data
    except KeyError:
        print(f"Error: Mission '{mission_name}' not found.")

# Example of usage
get_mission_data("Apollo 11")  # This will raise an error because the key doesn't exist
#3rd
valid_species = ["Homo sapiens", "Pan troglodytes", "Canis lupus", "Equus ferus caballus"]

def check_species(species_name):
    try:
        if species_name not in valid_species:
            raise ValueError(f"{species_name} is not a recognized species.")
        print(f"{species_name} is a valid species.")
    except ValueError as e:
        print(f"Error: {e}")

# Example of usage
check_species("Gorilla gorilla")  # This will trigger the error
#4rth
def calculate_evolutionary_rate(species_population, generations):
    try:
        if generations == 0:
            raise ZeroDivisionError("Generations cannot be zero when calculating evolutionary rates.")
        rate = species_population / generations
        return rate
    except ZeroDivisionError as e:
        print(f"Error: {e}")

# Example of usage
calculate_evolutionary_rate(1000, 0)  # This will trigger the error
#5th
genetic_sequences = {
    "Homo sapiens": "ATCGCGTACGTA",
    "Pan troglodytes": "ATCGGGTAGCTA",
}

def get_gene_sequence(species):
    try:
        sequence = genetic_sequences[species]
        return sequence
    except KeyError:
        print(f"Error: Gene sequence for {species} not found.")

# Example of usage
get_gene_sequence("Gorilla gorilla")  # This will raise an error because the key doesn't exist
# append functions
def append_to_protected_file(file_name, content):
    try:
        with open(file_name, 'a') as file:
            file.write(content + "\n")
            print("Content successfully appended to the file.")
    except PermissionError:
        print(f"Error: Permission denied to append to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
append_to_protected_file("protected_file.txt", "This is a test append.")  # Try appending to a protected file
#2nd
def append_multiple_lines(file_name, lines):
    try:
        with open(file_name, 'a') as file:
            for line in lines:
                file.write(line + "\n")
            print("Multiple lines successfully appended to the file.")
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied to append to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
lines_to_append = ["Additional Line 1", "Additional Line 2"]
append_multiple_lines("example.txt", lines_to_append)
#3rd
def append_to_file(file_name, content):
    try:
        with open(file_name, 'a') as file:
            file.write(content + "\n")
            print("Content successfully appended to the file.")
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied to append to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
append_to_file("example.txt", "This is appended content.")  # Append to 'example.txt'
#write functions
def write_multiple_lines(file_name, lines):
    try:
        with open(file_name, 'w') as file:
            for line in lines:
                file.write(line + '\n')
            print("Multiple lines successfully written to the file.")
    except PermissionError:
        print(f"Error: Permission denied to write to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
lines_to_write = ["Line 1", "Line 2", "Line 3"]
write_multiple_lines("example.txt", lines_to_write)
####
def write_to_file_with_disk_check(file_name, content):
    try:
        with open(file_name, 'w') as file:
            file.write(content)
            print("Content successfully written to the file.")
    except OSError as e:
        print(f"Error: {e.strerror} while writing to the file '{file_name}'.")
    except PermissionError:
        print(f"Error: Permission denied to write to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
write_to_file_with_disk_check("example.txt", "This content was written after checking disk space.")
###
def write_to_file(file_name, content):
    try:
        with open(file_name, 'w') as file:
            file.write(content)
            print("Content successfully written to the file.")
    except PermissionError:
        print(f"Error: Permission denied to write to the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
write_to_file("example.txt", "This is some new content.")  # Write to 'example.txt'
###read function
def read_empty_check(file_name):
    try:
        with open(file_name, 'r') as file:
            content = file.read()
            if not content:
                raise ValueError(f"The file '{file_name}' is empty.")
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except ValueError as e:
        print(e)  # Print if the file is empty
    except PermissionError:
        print(f"Error: Permission denied to read the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
read_empty_check("empty_file.txt")  # Check if 'empty_file.txt' is empty
##
def read_file_line_by_line(file_name):
    try:
        with open(file_name, 'r') as file:
            for line in file:
                print(line.strip())  # Strip newline characters
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied to read the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
read_file_line_by_line("example.txt")  # Try reading 'example.txt' line by line
##
def read_file(file_name):
    try:
        with open(file_name, 'r') as file:
            content = file.read()
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied to read the file '{file_name}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
read_file("example.txt")  # Try reading the file 'example.txt'

```
