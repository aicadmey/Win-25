# **Description**

This program performs an exploratory data analysis (EDA) on an asteroid dataset. The primary goal is to extract key insights from the data through comprehensive data inspection, summary statistics, and visualizations. The script includes the following steps:

**Data Loading & Inspection:**

The dataset is loaded from a CSV file. The program displays the data types, counts missing values and duplicates, and reports the overall shape of the dataset.

**Descriptive Statistics:**

It provides summary statistics for all numerical features to help understand central tendencies, dispersion, and data ranges.

**Visualization:**

Several bar charts are generated to highlight critical aspects of the data:

**Top 5 Largest Asteroids:**

Sorted by estimated minimum and maximum diameters.

**Top 5 Smallest Asteroids:**

 Based on estimated minimum and maximum diameters.

**Top 5 Fastest and Slowest Asteroids:**

 Determined by the relative velocity towards Earth.

**Top 5 Closest and Furthest Asteroids:**

Ranked by the miss distance.

**Top 5 Brightest and Dimmest Asteroids:**

Evaluated by absolute magnitude.

**Class Balance:**

 Visualizes the distribution of hazardous versus non-hazardous asteroids.



Unused libraries and models have been removed to maintain clarity and focus on the EDA process.

This streamlined script is perfect for quickly assessing the key characteristics of asteroid data and serves as a solid foundation for further analysis or model development.**

```python
# Define Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


from sklearn.feature_selection import SelectKBest
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import OneHotEncoder
from feature_engine.outliers import Winsorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load csv fiel
df = pd.read_csv('/content/drive/MyDrive/dataset asteroids/Asteroid_Dataset.csv')


print(df.dtypes)
print(df.isnull().sum().sort_values())
print(df.duplicated().sum())
print(df.shape)
print(df.describe())

# top 5 Samallest astoride

eda_minL = df.sort_values('est_diameter_min', ascending=False).head(5)
print(eda_minL)
eda_minL['est_diameter_min'].plot(kind='bar', ylabel='Estimated Min. Diameter (in KM)', color='red')
plt.show()

# Top 5 Largest Asteroids (in terms of Estimated Maximum Diameter)
eda_maxL = df.sort_values('est_diameter_max', ascending=False).head(5)
print(eda_maxL)
eda_maxL['est_diameter_max'].plot(kind='bar', ylabel='Estimated Max. Diameter (in KM)', color='darkred')
plt.show()

# Top 5 Smallest Asteroids (in terms of Estimated Minimum Diameter)
eda_minS = df.sort_values('est_diameter_min', ascending=True).head(5)
print(eda_minS)
eda_minS['est_diameter_min'].plot(kind='bar', ylabel='Estimated Min. Diameter (in KM)', color='blue')
plt.show()

# Top 5 Smallest Asteroids (in terms of Estimated Maximum Diameter)
eda_maxS = df.sort_values('est_diameter_max', ascending=True).head(5)
print(eda_maxS)

# Top 5 Fastest Asteroids
eda_rvL = df.sort_values('relative_velocity', ascending=False).head(5)
print(eda_rvL)
eda_rvL['relative_velocity'].plot(kind='bar', ylabel='Relative Velocity to Earth (in KPH)', color='yellow')
plt.show()

# Top 5 Slowest Asteroids
eda_rvS = df.sort_values('relative_velocity', ascending=True).head(5)
print(eda_rvS)
eda_rvS['relative_velocity'].plot(kind='bar', ylabel='Relative Velocity to Earth (in KPH)', color='orange')
plt.show()

# Top 5 Closest Asteroids
eda_mdS = df.sort_values('miss_distance', ascending=True).head(5)
print(eda_mdS)
eda_mdS['miss_distance'].plot(kind='bar', ylabel='Miss Distance (in KM)', color='red')
plt.show()

# Top 5 Furthest Asteroids
eda_mdL = df.sort_values('miss_distance', ascending=False).head(5)
print(eda_mdL)
eda_mdL['miss_distance'].plot(kind='bar', ylabel='Miss Distance (in KM)', color='orange')
plt.show()

# Top 5 Brightest Asteroids
eda_amL = df.sort_values('absolute_magnitude', ascending=False).head(5)
print(eda_amL)
eda_amL['absolute_magnitude'].plot(kind='bar', ylabel='Absolute Magnitude (in Watts Distance)', color='aqua')
plt.show()

# Top 5 Dimmest Asteroids
eda_amS = df.sort_values('absolute_magnitude', ascending=True).head(5)
print(eda_amS)
eda_amS['absolute_magnitude'].plot(kind='bar', ylabel='Absolute Magnitude (in Watts Distance)', color='teal')
plt.show()

# Class Balance
print("Class Balance: ", df['hazardous'].value_counts())
df['hazardous'].value_counts().plot(kind='bar')
plt.show()
```
