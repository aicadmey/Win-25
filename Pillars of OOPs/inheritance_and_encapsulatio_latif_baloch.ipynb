```python
class Planet:
    def __init__(self, name, size, composition, distance_from_star):
        self.name = name
        self.size = size  # e.g., small, medium, large
        self.composition = composition  # e.g., rocky, gaseous
        self.distance_from_star = distance_from_star  # in AU (Astronomical Units)

    def describe(self):
        return (f"{self.name} is a {self.size}-sized planet with a {self.composition} composition, "
                f"located {self.distance_from_star} AU from its star.")


class RockyPlanet(Planet):
    def __init__(self, name, distance_from_star):
        # Inherit size and composition but set them specifically for rocky planets
        super().__init__(name, "medium", "rocky", distance_from_star)


class GasGiant(Planet):
    def __init__(self, name, distance_from_star):
        # Gas giants inherit large size and gaseous composition
        super().__init__(name, "large", "gaseous", distance_from_star)


# Example usage:
earth = RockyPlanet("Earth", 1)
jupiter = GasGiant("Jupiter", 5.2)

print(earth.describe())  # Earth is a medium-sized planet...
print(jupiter.describe())  # Jupiter is a large-sized planet...



   class Rabbit:
    def __init__(self, fur_color, fitness_level):
        self.fur_color = fur_color  # e.g., "brown", "white"
        self.fitness_level = fitness_level  # e.g., 1 (low) to 10 (high)

    def __str__(self):
        return f"Rabbit with {self.fur_color} fur and fitness level {self.fitness_level}."


class BrownRabbit(Rabbit):
    def __init__(self, fitness_level):
        # Inherit characteristics of a Rabbit, but set fur color to brown
        super().__init__("brown", fitness_level)


class WhiteRabbit(Rabbit):
    def __init__(self, fitness_level):
        # White rabbits are more vulnerable, so they inherit a lower fitness level
        super().__init__("white", fitness_level)


# Simulating a generation of rabbits
generation = [
    BrownRabbit(fitness_level=8),  # Highly fit brown rabbit
    BrownRabbit(fitness_level=6),  # Less fit brown rabbit
    WhiteRabbit(fitness_level=3),  # Low fitness white rabbit
    WhiteRabbit(fitness_level=4),  # Low fitness white rabbit
]

# Print out the characteristics of the rabbits
for rabbit in generation:
    print(rabbit)

# Simulating the inheritance of traits and fitness in the next generation
# Assume that the fittest rabbits are more likely to reproduce
next_generation = [
    BrownRabbit(fitness_level=9),  # Offspring inherits better fitness
    BrownRabbit(fitness_level=7),  # Offspring inherits a bit better fitness
]

# Print out next generation
print("\nNext generation:")
for rabbit in next_generation:
    print(rabbit)



class Car:
    def __init__(self, make, model, year):
        # Public attributes
        self.make = make
        self.model = model
        # Private attribute
        self.__year = year

    # Public method to get the private attribute
    def get_year(self):
        return self.__year

    # Public method to set the private attribute
    def set_year(self, year):
        if year > 1885:  # Ensure valid year
            self.__year = year
        else:
            print("Invalid year")

# Creating an object of the Car class
my_car = Car("Toyota", "Corolla", 2020)

# Accessing public attribute
print(my_car.make)  # Toyota

# Trying to access private attribute directly (will raise an error)
# print(my_car.__year)  # AttributeError

# Accessing private attribute using getter method
print(my_car.get_year())  # 2020

# Modifying private attribute using setter method
my_car.set_year(2022)
print(my_car.get_year())  # 2022



 class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        # Private attribute for balance
        self.__balance = balance

    # Public method to get the balance
    def get_balance(self):
        return self.__balance

    # Public method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}")
        else:
            print("Deposit amount must be positive.")

    # Public method to withdraw money
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ${amount}")
        else:
            print("Invalid withdrawal amount.")

# Creating a BankAccount object
account = BankAccount("John Doe", 1000)

# Checking balance
print(account.get_balance())  # 1000

# Depositing money
account.deposit(500)  # Deposited: $500

# Withdrawing money
account.withdraw(200)  # Withdrawn: $200

# Checking balance after transactions
print(account.get_balance())  # 1300

# Trying to access private balance directly (will raise an error)
# print(account.__balance)  # AttributeError
```
